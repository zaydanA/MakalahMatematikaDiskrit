```python
import pandas as pd
from sklearn.datasets import load_iris

iris=pd.read_csv('.\\Dataset.csv')
iris
```

```python
iris.info()
iris.head()
```

DATA PREPROCESSING

```python
iris.drop('time',axis=1,inplace=True)
```

```python
iris.head()
```

```python
iris.replace('man',1,inplace=True) #1 adalah laki-laki
iris.replace('woman',0,inplace=True) #0  adalah perempuan
iris.head(20)
```

```python
x=iris[['height','shoe']]
y=iris['sex']

```

```python
iris.head()
```

```python
from sklearn.model_selection import train_test_split
 
# membagi dataset menjadi training dan testing 
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=123)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_model=DecisionTreeClassifier()

tree_model=tree_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred=tree_model.predict(X_test)
acc_skor= accuracy_score(y_pred,y_test)
print('Accuracy: ', acc_skor)
```

```python
#Memprediksi Hasil
print(tree_model.predict([[190.0, 45]])[0])
```

```python
from sklearn.tree import export_graphviz
export_graphviz(
    tree_model,
    out_file = "tree.dot",
    feature_names = ['height','shoe'],
    class_names=['woman','man'],
    rounded= True,
    filled =True)
```
